```python
#Invert Image Color

image1 = image.copy()
for col in range(image1.width):
    for row in range(image1.height):
        coords = (col, row)
        
        pixel = image1.getpixel(coords)
        
        #if red == 255, invert_red == 0 
        #if red == 254, invert_red == 1
        
        invert_colors = ()
        for color in pixel:
            invert_colors += (255 - color,)
            
        image1.putpixel(coords, invert_colors)
        
plt.imshow(image1)
plt.axis('off')
```

```python
#Black and White

image2 = image.copy()
for col in range(image2.width):
    for row in range(image2.height):
        coords = (col, row)
        pixel = image2.getpixel(coords)
        
        ave = sum(pixel)/3
        
        if ave < 128:
            ave = 0
        else:
            ave = 255
            
        image2.putpixel(coords, ((ave, )*3))
        
plt.imshow(image2)
plt.axis('off')
```

```python
#Greyscale

image3 = image.copy()
for col in range(image3.width):
    for row in range(image3.height):
        coords = (col, row)
        pixel = image3.getpixel(coords)
        
        ave = int(sum(pixel)/3)
            
        image3.putpixel(coords, ((ave, )*3))
        
plt.imshow(image3)
plt.axis('off')
```

```python
#Image Mirroring

image4 = image.copy()
for col in range(image4.width):
    for row in range(image4.height):
        coords = (col, row)
        
        mirr = (image4.width - col - 1, row)
        pixel = image.getpixel(mirr)
        
        image4.putpixel(coords, pixel)
        
plt.imshow(image4)
plt.axis('off')
```

```python
#Horizontal Flip

image5 = image.copy()
for col in range(image5.width):
    for row in range(image5.height):
        coords = (col, row)
        
        flip = (col, image5.height - row - 1)
        pixel = image.getpixel(flip)
        
        image5.putpixel(coords, pixel)
        
plt.imshow(image5)
plt.axis('off')
```

```python
#Contrast

image6 = image.copy()
for col in range(image6.width):
    for row in range(image6.height):
        coords = (col, row)
        pixel = image.getpixel(coords)
        
        contrast = ()
        for color in pixel
            if color < 128:
                color = int(color / 2)
            else: 
                color = color * 2
                
                if color > 255:
                    color = 255
                    
            contrast += (color,)
            
        contrast += (pixel[-1],)
        image6.putpixel(coords, contrast)
        
plt.imshow(image6)
plt.axis('off')
```

```python
#Image Detection

from skimage import io, feature
from skimage.color import rgb2gray

image_ = rgb2gray(io.imread("Lenna.png"))
```

```python
#Canny
edge_canny = feature.canny(image_)
plt.imshow(edge_canny)
plt.axis('off')
```

```python
#Filters: Prewitt
edge_prewitt = feature.prewitt(image_)
plt.imshow(edge_prewitt)
plt.axis('off')
```
